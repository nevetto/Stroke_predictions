Cleaning ever married column

```python
def clean_ever_married(x):
    if x == "Married":
        return "Yes"
    if x  == "Single":
        return "No"
    if x  == "Divorced":
        return "Other"
    return (x)

df4["ever_married"] = df4["ever_married"].apply(clean_ever_married)
```

```python
df4['ever_married'].unique()
```

cleaning work_type column

```python
df4['work_type'].unique()
```

```python
def clean_work_type(x):
    if x == "Government Job":
        return "Govt_job"
    if x  == "Never Worked":
        return "Never_worked"
    return (x)

df4["work_type"] = df4["work_type"].apply(clean_work_type)
```

```python
df4['work_type'].unique()
```

cleaning smoking_status column

```python
df4['smoking_status'].unique()
```

```python
def clean_smoking_status(x):
    if x == "Non-smoker":
        return "never smoked"
    if x  == "Formerly Smoked":
        return "formerly smoked"
    if x  == "Currently Smokes":
        return "smokes"
    return (x)

df4["smoking_status"] = df4["smoking_status"].apply(clean_smoking_status)
```

```python
df4['smoking_status'].unique()
```

cleaning stroke column

```python
df4["stroke"].unique()
```

```python
def clean_stroke(x):
    if x == "Stroke":
        return 1
     
    return int(x)

df4["stroke"] = df4["stroke"].apply(clean_stroke)
```
